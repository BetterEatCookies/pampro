# Autocalibration

Some imports we will need for the demo.

```python
from pampro import Time_Series, Channel, channel_inference, triaxial_calibration
from datetime import timedelta
```

The first thing we need to do is read in some triaxial data. For this experiment we will load from an exported activPAL CSV.

```python
ts = Time_Series.Time_Series("activPAL")
	
chans = Channel.load_channels("Z:/data/STVS/_data/activpal_data/activpal_data.csv", "activPAL")
x,y,z = chans[0],chans[1],chans[2]
ts.add_channels([x,y,z])
```

Before we perform the calibration, we can calculate Vector Magnitude (VM) and Euclidean Norm Minus One (ENMO) from the original data, which we will compare to later.

```python
vm = channel_inference.infer_vector_magnitude(x,y,z)
enmo = channel_inference.infer_enmo(vm)
```

We call the calibration method, passing it the raw X,Y,Z data channels. It returns the calibrated data channels, the parameters used to calibrate the data, an estimate of calibration error before and after, and information such as the number of still bouts identified and the quantity of time these bouts cover.

```python
cal_x, cal_y, cal_z, (cal_params), (results), (misc) = triaxial_calibration.calibrate(x,y,z)
```

```python
# The calibration values
print(cal_params)

# The time used to calibrate this device
print("Initial calibration error: " + str(results[0]))
print("Final calibration error: " + str(results[1]))
```

We can now repeat our calculation of VM and ENMO with the calibrated data.

```python
cal_vm = channel_inference.infer_vector_magnitude(cal_x, cal_y, cal_z)
cal_enmo = channel_inference.infer_enmo(cal_vm)
cal_vm.name = "VM_calibrated"
cal_enmo.name = "ENMO_calibrated"
```

Finally, we create an hourly summary of the VM and ENMO values, before and after calibration, and visualise the effect the calibration had. 

```python
ts.add_channels([vm, enmo, cal_vm, cal_enmo])


ts.add_channels(ts.piecewise_statistics(timedelta(hours=1), statistics={"VM":["mean"],"ENMO":["mean"],"VM_calibrated":["mean"],"ENMO_calibrated":["mean"]}, time_period=x.timeframe))

%matplotlib inline
ts.draw_experimental([["VM_mean", "VM_calibrated_mean"],["ENMO_mean", "ENMO_calibrated_mean"]], time_period=x.timeframe)
```

From the top chart, we can see that the calibrated VM is consistently closer to 1. From the bottom chart, we can see that calibrated ENMO follows a more sensible daily pattern, with a consistently near-zero ENMO overnight.
