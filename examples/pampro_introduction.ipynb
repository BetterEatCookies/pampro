```python
import numpy as np
from datetime import datetime, date, time, timedelta

from pampro import Time_Series, Channel, Bout, channel_inference

%matplotlib inline
```

```python
ts = Time_Series.Time_Series("Example Actiheart")

chans = Channel.load_channels("V:/P5_PhysAct/People/Tom/pampro/data/example_actiheart.txt", "Actiheart")
ts.add_channels(chans)

activity, heart = chans
```

```python
print(heart.timeframe)
```

We now have 2 channels of data in our time series object. We can visualise them like so:

```python
ts.draw_separate()
```

```python
stats = {"AH_Activity":["mean", "std"],"AH_ECG":["mean", "std"]}
```

```python
ts_output = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(minutes=15), stats)
ts_output.add_channels(simplified)
```

```python
ts_output.draw_separate(time_period=heart.timeframe)
```

We can also write time series objects to a file. 

```python
ts_output.write_channels_to_file("V:/P5_PhysAct/People/Tom/pampro/data/actiheart_summary.csv")
```

All we need to do is change the timedelta value to collapse the data to a different resolution - let's get daily data instead.

```python
ts_output2 = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(days=1), stats)
ts_output2.add_channels(simplified)
ts_output2.draw_separate()
```

A common way to analyse physical activity data is to slice it into relevant cutpoints, and count the time spent in them. This is built into the statistics dictionary. This code counts the minutes spent at different heart rate levels (40 to 80, 80 to 120, 120+) over hourly time windows, keeping the mean heart rate as a reference.

```python
stats = {"AH_ECG":["mean",[40,80],[80,120],[120,999]]}

ts_output3 = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(minutes=60), stats)
ts_output3.add_channels(simplified)
ts_output3.draw_separate(time_period=heart.timeframe)
```
