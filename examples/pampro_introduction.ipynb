# PAMPRO - Introduction.

```python
import numpy as np
from datetime import datetime, date, time, timedelta

from pampro import Time_Series, Channel, Bout, channel_inference, data_loading

%matplotlib inline
```

First we create a blank time series object, and read in some raw data as new channels. We tell the load function we want to load an Actiheart file, and where to find it; it returns a list of channels (activity and heart rate) that we can add to our time series object.

```python
ts = Time_Series.Time_Series("Example Actiheart")

chans = data_loading.load("V:/Projects/pampro/data/example_actiheart.txt", "Actiheart")
ts.add_channels(chans)

activity, heart = chans
```

We now have 2 channels of data in our time series object. We can visualise them like so:

```python
chart = ts.draw([["AH_Activity"], ["AH_ECG"]])
```

We can calculate statistics about our time series by defining a statistics dictionary. We name the channels we want to know about, and a list of statistics we want to derive about each. This calculates the mean and standard deviation for the activity and heart rate channels.

```python
stats = {"AH_Activity":[("generic", ["mean", "std"])],"AH_ECG":[("generic", ["mean", "std"])]}
```

We can use this statistics dictionary to request piecewise statistics of the channels, broken down into any time window we like by specifying a timedelta object. Each statistic becomes a channel of data itself, which we can add to another time series object.

```python
ts_output = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(minutes=15), stats)
ts_output.add_channels(simplified)
```

We have now calculated the mean and standard deviation of those two channels over 15 minute time windows. This is now a time series object itself, and we can visualise our four derived statistics.

```python
chart = ts_output.draw([["AH_ECG_mean"],["AH_ECG_std"],["AH_Activity_mean"],["AH_Activity_std"]], time_period=heart.timeframe)
```

We can also write time series objects to a file. This will write the data of every channel to the specified file in long format.

```python
ts_output.write_channels_to_file("V:/Projects/pampro/data/actiheart_summary.csv")
```

All we need to do is change the timedelta value to collapse the data to a different resolution - let's get daily data instead.

```python
ts_output2 = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(days=1), stats)
ts_output2.add_channels(simplified)
chart = ts_output2.draw([["AH_ECG_mean"],["AH_ECG_std"],["AH_Activity_mean"],["AH_Activity_std"]], time_period=heart.timeframe)
```

A common way to analyse physical activity data is to slice it into relevant cutpoints, and count the time spent in them. This is built into the statistics dictionary. This code counts the minutes spent at different heart rate levels (40 to 80, 80 to 120, 120+) over hourly time windows, keeping the mean heart rate as a reference.

```python
stats = {"AH_ECG":[("generic", ["mean"]),("cutpoints", [[40,80],[80,120],[120,999]])]}

ts_output3 = Time_Series.Time_Series("Output")
simplified = ts.piecewise_statistics(timedelta(minutes=60), stats)
ts_output3.add_channels(simplified)
chart = ts_output3.draw([["AH_ECG_mean"], ["AH_ECG_40_80"], ["AH_ECG_80_120"],["AH_ECG_120_999"] ], time_period=heart.timeframe)
```
