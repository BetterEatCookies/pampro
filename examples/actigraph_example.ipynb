```python
from datetime import datetime, date, time, timedelta
from pampro import data_loading, Time_Series, Channel, channel_inference, Bout
%matplotlib inline
```

Firstly, load the supplied example Actigraph data.

```python
counts, header = data_loading.load("V:/Projects/pampro/data/example_actigraph.DAT", "Actigraph", datetime_format="%m/%d/%Y")
```

That function returned a channel of count data (a Channel object), and some useful information from the header (a Python dictionary). We start by printing out the header data.

```python
for key,value in header.items():
    print(key,value)
    
ts = Time_Series.Time_Series("Actigraph")
ts.add_channel(counts)
```

Let's draw the count signal and see what it looks like:

```python
chart = ts.draw([["AG_Counts"]])
chart.axes[0].set_ylim(-5, 10000)
```

Get a list of bouts where the monitor was & wasn't worn.

```python
nonwear_bouts, wear_bouts = channel_inference.infer_nonwear_actigraph(counts, zero_minutes=timedelta(minutes=90))
```

We will use these bouts to exclude nonwear periods from the analysis. We can superimpose these bouts on the original chart, and show where the nonwear occurs:

```python
counts.add_annotations(nonwear_bouts)
chart = ts.draw([["AG_Counts"]])
chart.axes[0].set_ylim(-5, 10000)
```

It looks like this person wore the monitor over night on the 3rd day, so that isn't shaded as nonwear. Now we can use the list of wear bouts to get another list of days that we consider valid & invalid:

```python
invalid_bouts, valid_bouts = channel_inference.infer_valid_days(counts, wear_bouts)
```

Again, we can superimpose the invalid bouts to see which days we won't include.

```python
for ib in invalid_bouts:
    ib.draw_properties = {'lw':0, 'alpha':0.8, 'facecolor':[0.78431,0.1,0.1]}
    
counts.add_annotations(invalid_bouts)
chart = ts.draw([["AG_Counts"]])
chart.axes[0].set_ylim(-5, 10000)
```

We're going to fill sections of the count signal with negative data, to exclude it from the analysis. Where we found nonwear, we'll set it to -1, and where we defined the day as invalid, we'll set it to -2.

```python
counts.fill_windows(nonwear_bouts, fill_value=-1)
counts.fill_windows(invalid_bouts, fill_value=-2)
```

Let's see the resulting channel:

```python
chart = ts.draw([["AG_Counts"]])
chart.axes[0].set_ylim(-5, 10000)
```

We can see the invalid data has been set to -2, and more subtly, the nonwear zeroes have been set to -1. 

Now we can start applying cutpoints to the data - we can count the negative numbers to know how much nonwear & invalid time there was, and use generic cutpoints (0 to 99, etc) to define sedentary time.

```python
stats = {"AG_Counts": [("cutpoints", [[-2,-2],[-1,-1],[0,99999],[0,99],[100,2999],[3000,99999]])]}
summary_results = ts.summary_statistics(statistics=stats)
```

Create a time series object, put the summary level results in it, and write them to a file.

```python
ts_output = Time_Series.Time_Series("Output")
ts_output.add_channels(summary_results)
ts_output.write_channels_to_file("V:/Projects/example_output.csv")
```

Alternatively, we produce an hourly analysis of the same statistics. That could also be written to a file in long-format. 

```python
hourly_results = ts.piecewise_statistics(timedelta(hours=1), statistics=stats)
ts_output = Time_Series.Time_Series("Output")
ts_output.add_channels(hourly_results)
ts_output.write_channels_to_file("V:/Projects/example_output.csv")
```

Piecewise statistics data also comes out as PAMPRO channel objects. Which means we can visualise it the same way:

```python
chart = ts_output.draw([["AG_Counts_0_99", "AG_Counts_100_2999", "AG_Counts_3000_99999", "AG_Counts_0_99999"]], time_period=counts.timeframe)
chart.axes[0].set_ylim(-1, 61)
```
